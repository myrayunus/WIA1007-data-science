```python
#4. Create a random numpy array of size 6x3 and transpose it
import numpy as np

def transpose_array():
  original_array = np.random.rand(6,3)
  transposed_array = original_array.T
  print("Original array: ")
  print(original_array)
  print(" Transposed array:  ")
  print(transposed_array)

transpose_array()
```

```python
#5. Create two random numpy array of size n x p and p x m and then compute the dot product
import numpy as np
n=int(input("Enter n:"))
p=int(input("Enter p:"))
m=int(input("Enter m:"))

def dot_product(n,p,m):
  array_A = np.random.rand(n,p)
  array_B = np.random.rand(p,m)
  dot_product = array_A @ array_B
  return dot_product

dot_product = dot_product(n,p,m)
print(dot_product)
```
